```python
from sklearn.preprocessing import MinMaxScaler
# Select relevant features (numerical audio features + genre)
features = ['valence', 'danceability', 'energy', 'acousticness',
            'instrumentalness', 'liveness', 'speechiness', 'playlist_genre']

# Normalize numerical features (0-1 scaling)
scaler = MinMaxScaler()
df[features[:-1]] = scaler.fit_transform(df[features[:-1]])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a "genre + features" representation for each song
df['genre_features'] = df['playlist_genre'] + " " + \
                       df[features[:-1]].astype(str).apply(' '.join, axis=1)

# Vectorize using TF-IDF (or just use numerical features directly)
tfidf = TfidfVectorizer()
song_vectors = tfidf.fit_transform(df['genre_features'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_songs(track_name, n=50):
    # Find the song's index
    idx = df[df['track_name'] == track_name].index[0]

    # Calculate cosine similarity
    sim_scores = cosine_similarity(song_vectors[idx], song_vectors).flatten()

    # Get top-N most similar songs (excluding itself)
    similar_songs = sorted(list(enumerate(sim_scores)), key=lambda x: x[1], reverse=True)[1:n+1]

    # Return recommendations
    return df.iloc[[i[0] for i in similar_songs]][['track_name', 'playlist_genre']]
```

```python
print(recommend_songs("APT."))
```
